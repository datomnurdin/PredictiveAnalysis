Before going any further, there are a couple of null values in the data that we need to clean up.

```python
fullData[fullData.isnull().any(axis=1)]
```

For the sake of simplicity, just fill in the value with the common value (since that is the most common value):

Product: Warehousing

Cust Dec Date: 30/03/2017

Go Live Date: 03/07/2017

BCA_Status: Approved

Payment Terms: 2-3 Years

```python
fullData = fullData.fillna({"Product": "Warehousing"})
fullData = fullData.fillna({"Cust Dec Date": "30/03/2017"})
fullData = fullData.fillna({"Go Live Date": "03/07/2017"})
fullData = fullData.fillna({"BCA_Status": "Approved"})
fullData = fullData.fillna({"Payment Terms": "2-3 Years"})
```
